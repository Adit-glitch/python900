```python
import matplotlib.pyplot as plt

```

```python
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
inflation = [4, 5.6, 5.3, 5.5, 6, 7]
```

```python
plt.plot(x, inflation)
```

```python
plt.plot(x, inflation, 'o-', color = 'red') # 'o-' is for dot
```

```python
plt.plot(x, inflation, 'o-', color = 'red') # 'o-' is for dot
plt.xlabel('Month')
plt.ylabel('inflation')
plt.show()
```

# bar

```python
party = ['RPP', 'RSP', 'Maobadi', 'Congress', 'CPN-UML']
seat = [14, 24, 32, 89, 79]
```

```python
plt.bar(x = party, height = seat, color = ['red', 'yellow', 'orange', 'blue', 'black'])
plt.show()
```

```python
plt.pie(seat, labels = party, autopct = '%.1f%%')
plt.show()
```
